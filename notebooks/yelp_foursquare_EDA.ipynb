```python
import requests

#STATION 1

url = "https://api.yelp.com/v3/businesses/search"
key = yelp_key
parameters = {'latitude':25.729275,
              'longitude':-80.2415815,
              'radius' : 1000,
              'limit' : 50}
headers = {
    'Authorization': f'Bearer {key}'
}
response = requests.get(url, headers=headers, params=parameters)
response.status_code
YELP_station1 =response.json()['businesses']
YELP_station1 = pd.json_normalize(YELP_station1)
YELP_station1.head()
categories_df = pd.json_normalize(YELP_station1['categories'])
#categories_df.head()
```
