Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python

import pandas as pd

# Load the DataFrame from the Miami Bikes pickle file
df1 = pd.read_pickle('samples_3_Miami_Bikes.pk')
df1.head()
```

```python
#Using 3 Stations
station1_bikes = df1.loc[0:0]
station2_bikes = df1.loc[1:1]
station3_bikes = df1.loc[2:2]
```

```python
# Load the DataFrame from YELP_Miami_Bike station 1 pickle file
station1_yelp = pd.read_pickle('YELP_station_Miami_Bikes_Station1.pk')
station2_yelp = pd.read_pickle('YELP_station_Miami_Bikes_Station2.pk')
station3_yelp = pd.read_pickle('YELP_station_Miami_Bikes_Station3.pk')
```

```python
#station1combined_df 
# Repeat the single-row DataFrame to match the number of rows in df2
station1_bikes_repeated = pd.concat([station1_bikes] * len(station1_yelp), ignore_index=True)

# Concatenate the two DataFrames horizontally
station1combined_df= pd.concat([station1_bikes_repeated, station1_yelp], axis=1)
#station1combined_df
```

```python
#station2combined_df

# Repeat the single-row DataFrame to match the number of rows in df2
station2_bikes_repeated = pd.concat([station2_bikes] * len(station2_yelp), ignore_index=True)

# Concatenate the two DataFrames horizontally
station2combined_df= pd.concat([station2_bikes_repeated, station2_yelp], axis=1)
#station2combined_df
```

```python
#station3combined_df

# Repeat the single-row DataFrame to match the number of rows in df2
station3_bikes_repeated = pd.concat([station3_bikes] * len(station3_yelp), ignore_index=True)

# Concatenate the two DataFrames horizontally
station3combined_df= pd.concat([station3_bikes_repeated, station3_yelp], axis=1)
#station3combined_df
```

```python
#ALL STATIONS COMBINED

All_3_Miami_Stations_Combined = pd.concat([station1combined_df,station2combined_df,station3combined_df], axis=0, ignore_index=True)
All_3_Miami_Stations_Combined.head(1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
review_counts = [19838, 15907, 18702]
ratings = [4.03, 4.06, 4.04]
bike_stations = ['605 - Virginia St & Florida Ave', '606 - S Bayshor Dr & Aviation Ave', '608 - City Hall']

# Set the width of each bar
bar_width = 0.15

# Create an array of indices for the x-axis locations
indices = np.arange(len(bike_stations))

# Create the figure and axes
fig, ax1 = plt.subplots()

# Plot the review counts
ax1.bar(indices, review_counts, bar_width, color='blue', label='Review Count')

# Set labels and title for the first y-axis
ax1.set_ylabel('Review Count')
ax1.set_title('Review Count and Rating by Bike Stations')

# Create a secondary y-axis
ax2 = ax1.twinx()

# Plot the ratings
ax2.plot(indices + bar_width / 2, ratings, marker='o', color='red', label='Average Rating')

# Set labels for the second y-axis
ax2.set_ylabel('Average Ratings')

# Set the x-axis tick positions and labels
ax1.set_xticks(indices)
ax1.set_xticklabels(bike_stations, rotation=45)

# Adjust the layout and spacing
fig.tight_layout()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Define the color palette
colors = ['blue', 'green', 'red']

# Extract the category columns
category_columns = ['Category 1']
category_data = station1combined_df[category_columns]

fig = plt.figure(figsize=(9.5, 4))
# Plot the histogram for each category
for i, category in enumerate(category_columns):
    plt.hist(category_data[category], bins=80, color=colors[i], label=category)

# Set labels and title
plt.xlabel('POI Florida Avenue')
plt.ylabel('Frequency')
plt.title('POI Categories Florida Avenue')
#plt.legend()
plt.xticks(rotation=75)
# Display the plot
plt.show()
```
